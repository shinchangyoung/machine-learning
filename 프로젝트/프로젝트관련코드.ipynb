```python
import pandas as pd
import re

# CSV 파일 불러오기
df = data  # 이미 data라는 변수에 데이터프레임이 있다고 가정

def clean_name(name):
    if pd.isna(name):
        return name
    # 괄호 내용 제거
    name = re.sub(r'\(.*?\)', '', name)
    # 공백 이후 내용 제거
    name = name.split()[0]
    return name.strip()

# '상권_코드_명' 컬럼을 직접 수정
df['상권_코드_명'] = df['상권_코드_명'].apply(clean_name)
```

```python

import pandas as pd
from geopy.geocoders import Nominatim
import time

# CSV 파일 경로 (예: 'data.csv')

# CSV 파일 읽기
df = data

# '상권코드명'에서 '메디힐병원'이 포함된 행 필터링
filtered_df = df[df['상권_코드_명'].str.contains('메디힐병원', na=False)]

# 병원 이름으로 좌표 검색
geolocator = Nominatim(user_agent="geoapiExercises")

# 결과 리스트 초기화
results = []

for index, row in filtered_df.iterrows():
    place_name = row['상권_코드_명']

    try:
        location = geolocator.geocode(place_name)
        time.sleep(1)  # 요청 간 간격을 주어야 블럭되지 않음

        if location:
            latitude = location.latitude
            longitude = location.longitude
        else:
            latitude = None
            longitude = None

        # 기존 컬럼 + 위도, 경도 추가
        row_data = row.to_dict()
        row_data['위도'] = latitude
        row_data['경도'] = longitude

        results.append(row_data)

    except Exception as e:
        print(f"에러 발생: {e}")

# 결과를 데이터프레임으로 변환
result_df = pd.DataFrame(results)

# 결과 출력
print(result_df)

# CSV로 저장하고 싶다면:
# result_df.to_csv('메디힐병원_결과.csv', index=False)
```

```python
import pandas as pd
!pip install geopy
from geopy.geocoders import Nominatim

data=pd.read_csv('/content/신차영 (1).csv')

def get_coordinates(location_name):
    # 사용자 에이전트 이름 설정 (필수)
    geolocator = Nominatim(user_agent="location_finder_app")

    try:
        # 위치 정보 조회
        location = geolocator.geocode(location_name)

        if location:
            # 위도, 경도 반환
            return {
                "latitude": location.latitude,
                "longitude": location.longitude
            }
        else:
            return {"error": "위치를 찾을 수 없습니다."}

    except Exception as e:
        return {"error": f"오류 발생: {str(e)}"}

# 사용 예시
if __name__ == "__main__":
    res=[]
    for i in range(0,len(data)):
      res.append(get_coordinates(data['상권_코드_명'][i]))
      print(i/len(data)*100)
```

```python
import pandas as pd
import numpy as np
import math
from sklearn.cluster import KMeans

file = pd.read_csv('/content/역_버스정류장_위치_정보.csv', encoding='cp949')

latitudes = file['위도'].to_numpy()
longitudes = file['경도'].to_numpy()
coords = np.vstack((latitudes, longitudes)).T

mean_lat = latitudes.mean()
lat_km = coords[:, 0] * 111.0
lon_km = coords[:, 1] * 111.0 * np.cos(np.radians(mean_lat))
coords_km = np.vstack((lat_km, lon_km)).T
coords_km
```

```python
test=[]
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
for k in range(2,5001):
 k_model = KMeans(n_clusters=k, random_state=42)
 labels = k_model.fit_predict(coords_km)
 score = silhouette_score(coords_km, labels)
 test.append(score)
 print(k/10001*100)

```
