```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[2,4,6,8]


#label이름
plt.xlabel('x-label')
plt.ylabel('y-label')

#plt.plot(x)
#plt.plot(y)
#이렇게 두개로 묶어서 하는것도 가능하다!
#plot는 죄표평면에 x축과 y축의 값을 나타내는 코드
plt.plot(x,y)

#그래프를 실질적으로 보여주는 코드
plt.show()
```

```python
#(x,y)로 나타내는 방법
#plt.plot([1,2,3,4],[2,4,6,8], label = 'square')

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]
y2 = [1, 3, 5, 7]  # y2의 길이를 x와 동일하게 4로 맞춤

plt.plot(x, y1, label="y = 2x")
plt.plot(x, y2, label="y = x + 1")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("TITLE")

#x,y축의 첫하고 끝의 길이 선정 코드
#예로 plt.xlim([8,10])를 코드는 8에서 10까지의 소수점 부분을 x축에 표기
plt.xlim([0,10])
plt.ylim([0,10])

#범례를 사용하기 위한 코드는 legend: 범례
#loc은 범례에 위치를 선정해주고 ncol은 범례의 행렬 즉 열을 선정해준다
#**"upper" (위쪽)**
plt.legend(loc = 'upper left', ncol = 2)  # 범례 추가
plt.show()
```

```python
#axis([xmax,xmin,ymax,ymin])
#축 범위 지정


plt.plot(x, y1, '--', label="y = 2x")
plt.plot(x, y2, label="y = x + 1")


plt.axis([0,15,0,15])

#loc: Location (위치)
#ncol: Number of Columns (열 개수)

plt.legend(loc = 'upper right', ncol=2)
plt.show()

```

```python
import matplotlib.pyplot as plt

y1 = [2,3,4,5]
y2 = [1,3,5,7]

plt.plot([4,5,6], 's')
plt.plot([5,6,7], 'b')
plt.plot([7,8,9], color = 'violet', label = '[7,8,9] = violet')

plt.plot(y1, '--', label = 'y1=blue',)
plt.plot(y2, '-',label = 'y2=orange')

plt.axis([0,10,0,10])

plt.legend(loc = 'upper left', ncol=2)
plt.show()
```

그래프 꾸미기

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,4,6,8]

a = [8,9,10,11]
b = [2,6,10,14]

plt.axis([0,15,0,15])
plt.plot(x, color = 'red', label = 'x = red')
plt.plot(y, color = 'blue', label = 'y = blue')

plt.plot(a, color = 'yellow', label = 'a')
plt.plot(b, color = 'orange', label = 'b')


#그래프의 제목을 꾸미기
#plt.title(), x, ylabel()에 사용이 가능
#fontsize, fontweight, color, backgroundcolor 등등
plt.title('graph', fontsize = 20, pad = 50, fontweight = 'bold', color = 'violet', backgroundcolor = 'lightgray')
plt.xlabel('xaxis' , color = 'red', fontsize = 12)
plt.ylabel('yaxis' , color = 'blue', fontsize = 18, fontweight = 'bold')

plt.legend(loc = 'upper center', ncol = 2)
plt.show()
```

슈퍼타이틀 suptitle()

```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.figure(figsize=(6, 4))


#plt.subplot(1, 2, 2)의 의미
#plt.subplot(nrows, ncols, index)
#nrows: 행(row)의 개수
#ncols: 열(column)의 개수
#index: 몇 번째 서브플롯인지 지정 (왼쪽 → 오른쪽, 위 → 아래 순서)

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("Graph 1")

plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title("Graph 2")



plt.suptitle("Main Title", fontsize=14, fontweight='bold')

# 1) 기본 설정 (위쪽 여백 5%)
plt.tight_layout(rect=[0, 0, 1, 0.95])

# 2) 위쪽 여백을 줄이면 제목이 겹칠 가능성 있음
# plt.tight_layout(rect=[0, 0, 1, 0.9])

# 3) 서브플롯을 오른쪽으로 조금 좁게 조정
# plt.tight_layout(rect=[0.05, 0, 0.95, 0.95])

plt.show()
```

xtixks(), yticks()함수 사용

```python
import matplotlib.pyplot as plt

x = [2,4,6]
WorldCup = ['2000year', '2018year', '2022year']
Values = [500, 200, 300]

plt.ylim([0,600])
plt.bar(x, Values, color = ['r', 'b', 'g'], label = ['2002 = r', '2018 = b', '2022 = g'])


plt.title('best korea worldcup!', loc = 'center', pad = 50, fontsize = 20, color = 'b', fontweight = 'bold')
plt.xlabel('year', color = 'violet', fontsize = 15)
plt.ylabel('values', color = 'violet', fontsize = 15)
plt.legend(loc = 'upper center', ncol = 3)
plt.xticks(x,WorldCup)
plt.show()
```
