```python
import matplotlib.pyplot as plt
import numpy as np

#  
x = np.linspace(1, 10, 100)
y = np.exp(x)

#  
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('Linear Scale')

#  
plt.subplot(2, 2, 2)
plt.plot(x, y)
plt.yscale('log')
plt.title('Logarithmic Scale (y-axis)')

# - 
plt.subplot(2, 2, 3)
plt.plot(x, y)
plt.xscale('log')
plt.title('Log-Linear Scale (x-axis)')

# - 
plt.subplot(2, 2, 4)
plt.plot(x, y)
plt.yscale('log')
plt.xscale('linear')
plt.title('Linear-Logarithmic Scale')

#    
plt.tight_layout()
plt.show()


#  :   
# :   
```
