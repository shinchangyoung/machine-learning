```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(1, 10, 100)
y = np.exp(x)

# 선형 스케일
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('Linear Scale')

# 로그 스케일
plt.subplot(2, 2, 2)
plt.plot(x, y)
plt.yscale('log')
plt.title('Logarithmic Scale (y-axis)')

# 로그-선형 스케일
plt.subplot(2, 2, 3)
plt.plot(x, y)
plt.xscale('log')
plt.title('Log-Linear Scale (x-axis)')

# 선형-로그 스케일
plt.subplot(2, 2, 4)
plt.plot(x, y)
plt.yscale('log')
plt.xscale('linear')
plt.title('Linear-Logarithmic Scale')

# 서브 그래프 간격 조절
plt.tight_layout()
plt.show()


# 선형스케일 : 일정 범위에서 증가
#로그스케일 : 배수 범위에서 증가
```
